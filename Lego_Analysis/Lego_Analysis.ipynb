```python
import pandas as pd

df = pd.read_csv('datasets/lego_sets.csv')

theme = pd.read_csv('datasets/parent_themes.csv')

theme.head()
```

```python
df.head()
```

What percentage of all licensed sets ever released were star wars themed

```python
merged_df = df.merge(theme, left_on='theme_name', right_on='name')
merged_df.drop(columns='name_y', inplace=True)
merged_df.shape
```

```python
licensed = merged_df[merged_df['is_licensed']]
licensed
```

```python
star_wars = licensed[licensed['parent_theme'] == 'Star Wars']
star_wars.head()
```

```python
the_force = star_wars.shape[0] / licensed.shape[0]
```
